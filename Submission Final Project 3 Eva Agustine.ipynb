```python
# Data Source
import pandas as pd
df = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')
df.head()
```

```python
# Pre processing data train
#Imputing missing value
mode_kelas_pekerja = df['Kelas Pekerja'].mode()
mode_pekerjaan = df['Pekerjaan'].mode()

df['Kelas Pekerja'] = df['Kelas Pekerja'].replace(['?'], mode_kelas_pekerja)
df['Pekerjaan'] = df['Pekerjaan'].replace(['?'],mode_pekerjaan)

#Encoding data
df= pd.get_dummies(df, columns=['Kelas Pekerja', 'Status Perkawinan', 'Jenis Kelamin', 'Pekerjaan'])
pendidikan_dict = {'1st-4th' : 0 , '5th-6th' : 1, 'SD' : 2, '7th-8th': 3, '9th' : 4,  '10th' : 5, '11th' :6, 'SMA' :7, 
                   'D3' : 8, 'D4' : 9,  '12th' :10, 'Pendidikan Tinggi' : 11, 'Sarjana' :12,'Master':13, 'Doktor':14,
                   'Sekolah Professional' : 15}
df['Pendidikan'] = df['Pendidikan'].replace(pendidikan_dict)
gaji_dict = {'<=7jt' : 0 ,  '>7jt' : 1}
df['Gaji'] = df['Gaji'].replace(gaji_dict)
df.head()
```

```python
#Centering and Scaling Data train
from sklearn.preprocessing import StandardScaler

x = df.drop(['Gaji', 'id'], axis=1)
y = df['Gaji']
#stdScalar = StandardScaler()
#dataScale = stdScalar.fit_transform(x)

#colnames = list(df.columns.drop(['Gaji', 'id'])) 
#dfscale = pd.DataFrame(dataScale, columns=colnames)
```
